```python
#Divide the dataset to 5 clients
import os, shutil, random

img_dir = '/content/VisDrone_Subset/images/train'
label_dir = '/content/VisDrone_Subset/labels/train'

for i in range(5):
    os.makedirs(f'/content/client{i}/images', exist_ok=True)
    os.makedirs(f'/content/client{i}/labels', exist_ok=True)

all_imgs = sorted([f for f in os.listdir(img_dir) if f.endswith(('.jpg', '.png'))])
random.shuffle(all_imgs)
split_size = len(all_imgs) // 5

for i in range(5):
    subset = all_imgs[i*split_size : (i+1)*split_size]
    for img in subset:
        label = img.rsplit('.', 1)[0] + '.txt'
        shutil.copy(os.path.join(img_dir, img), f'/content/client{i}/images/{img}')
        shutil.copy(os.path.join(label_dir, label), f'/content/client{i}/labels/{label}')
```

```python
#Create yaml file with 3 classes
names = ['car', 'van', 'truck']
nc = len(names)

for i in range(5):
    yaml_text = f"""
path: /content/client{i}
train: images
val: /content/VisDrone_Subset/images/val
nc: {nc}
names: {names}
"""
    with open(f"/content/client{i}.yaml", "w") as f:
        f.write(yaml_text.strip())
```

```python
import torch
import flwr as fl
from ultralytics import YOLO
from flwr.common import parameters_to_ndarrays
import os
from ultralytics.nn.modules.head import Detect

PHASE_ID = 1

CLIENT_PHASES = {
    0: [0, 1],
    1: [2, 3],
    2: [4, 0]
}

initial_model_path = "/content/drive/MyDrive/fed_model0.pt"
project = "/content/runs/shiyan/"
name = "ceshi"
save_model= project + name

class FedYOLOClient(fl.client.NumPyClient):
    def __init__(self, client_id: int, model_path: str):
        self.client_id = client_id
        self.model = YOLO(model_path)
        self.yaml_path = f"/content/client{client_id}.yaml"
        self.image_dir = f"/content/client{client_id}/images"

    def get_parameters(self, config=None):
        return [val.cpu().numpy() for val in self.model.model.state_dict().values()]

    def set_parameters(self, parameters):
        model_keys = list(self.model.model.state_dict().keys())
        state_dict = self.model.model.state_dict()
        for k, p in zip(model_keys, parameters):
            tensor = torch.tensor(p)
            if state_dict[k].shape == tensor.shape:
                state_dict[k] = tensor
            else:
                print(f"Ignore shape mismatch value：{k}")
        self.model.model.load_state_dict(state_dict, strict=False)

    def fit(self, parameters, config=None):
        self.set_parameters(parameters)
        self.model.train(
            data=self.yaml_path,
            epochs=50,
            imgsz=640,
            batch=8,
            device=0,
            verbose=False,
            workers=0,
            project=project,
            name=name,
            exist_ok=True,
        )
        return self.get_parameters(), len(os.listdir(self.image_dir)), {}

    def evaluate(self, parameters, config=None):
        self.set_parameters(parameters)
        metrics = self.model.val(data=self.yaml_path, verbose=False)
        return float(metrics.box.map50), len(os.listdir(self.image_dir)), {}
```

```python
def client_fn(cid: str) -> fl.client.NumPyClient:
    cid_int = int(cid)
    if cid_int in CLIENT_PHASES[PHASE_ID]:
        return FedYOLOClient(
            client_id=cid_int,
            model_path=initial_model_path
        )
    else:
        class DummyClient(fl.client.NumPyClient):
            def __init__(self):
              self.model = YOLO(initial_model_path)
              print(f"DummyClient initialized successfully，the model variable: {len(self.model.model.state_dict())}")

            def get_parameters(self, config=None):
                return [val.cpu().numpy() for val in self.model.model.state_dict().values()]

            def fit(self, parameters, config=None):
                return self.get_parameters(), 0, {}
            def evaluate(self, parameters, config=None):
                return 0.0, 0, {}
        print(f"Skip Client{cid_int}")
        return DummyClient()

class SaveModelStrategy(fl.server.strategy.FedAvg):
    def aggregate_fit(self, server_round, results, failures):
        aggregated_parameters, _ = super().aggregate_fit(server_round, results, failures)
        if aggregated_parameters is not None:
            ndarrays = parameters_to_ndarrays(aggregated_parameters)

            model = YOLO(initial_model_path)
            state_dict = model.model.state_dict()
            new_state_dict = {}
            for k, v in zip(state_dict.keys(), ndarrays):
                if state_dict[k].shape == torch.tensor(v).shape:
                    new_state_dict[k] = torch.tensor(v)
                else:
                  print(f"Ignore variable shape mismatch: {k}")
            model.model.load_state_dict(new_state_dict, strict=False)

            save_path = f"{save_model}/weights/fed_model_round{server_round}.pt"
            model.save(save_path)
            print(f"The model saved successfully: {save_path}.pt")
        return aggregated_parameters, {}
```

```python
strategy = SaveModelStrategy(
    fraction_fit=1.0,
    fraction_evaluate=1.0,
    min_fit_clients=2,
    min_evaluate_clients=0,
    min_available_clients=2,
)

fl.simulation.start_simulation(
    client_fn=client_fn,
    num_clients=5,
    config=fl.server.ServerConfig(num_rounds=2),
    strategy=strategy,
    ray_init_args={"include_dashboard": False}
)
```
